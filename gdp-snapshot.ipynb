```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Sep 23 15:42:46 2020

@author: kalyan

Proof of Concept to retrieve data via DBNOMICS API
"""

import pandas as pd  # pandas

import plotly.io as pio
from plotly.offline import plot
import plotly.express as px
import plotly.graph_objects as go

from dbnomics import fetch_series, fetch_series_by_api_link  # Install dbnomics to retrieve macro timeseries


#pio.renderers.default = "browser"

def gen_gdpsnapshot(cons_ind_list, max_rows=100):
    '''
    Future improvements include shortening code by using loops for quandrants.
    #quadrant = [1,2,3,4]

    '''
    cons_ind = list(cons_ind_list) # Consumption Data Indicators list
    #print(type(cons_ind))
    #print(len(cons_ind))
    tracelist_q2 = []
    # Private Final Consumption Expenditure (Constant(2010) US Dollars) 
    # Figure Layout & Traces and Figure definition
    layout_q2 = go.Layout(height=500, width=700,
            title = dict(text="Consumption (Constant(2010) US Dollars)"),
            xaxis=dict(title="Year"),
            yaxis=dict(title="Consumption (US $)"),
            grid=None
            )
    for i in range(len(cons_ind)):
        df_i = fetch_series(cons_ind[i]) 
        #print(df_i['series_code'].unique())
        traces_q2_i = go.Scatter(x=df_i['period'], y=df_i['value'], name=str(df_i['series_code'].unique()))
        tracelist_q2.append(traces_q2_i)
        i=i+1

    fig_quad_2 = go.Figure(data = tracelist_q2, layout=layout_q2)
    fig_quad_2.update_layout(hovermode='x unified')
    #fig_quad_2.show()
    return fig_quad_2 
```
