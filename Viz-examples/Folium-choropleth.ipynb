```python
# Import libraries
import pandas as pd
import folium
import os as os 

# Load the shape of Indian States
# Find the original file here: https://github.com/python-visualization/folium/tree/master/examples/data
# You have to download this file and set the directory where you saved it
ind_state_basemap = os.path.join('/Users/kalyan/OneDrive-IIMV/OneDrive - IIM VIZAG/Python-Exercise-KK/maps-master/States/', 'Admin2.json')
 
# Load the unemployment value of each state
# Find the original file here: https://github.com/python-visualization/folium/tree/master/examples/data
ind_state_filepath = os.path.join('/Users/kalyan/OneDrive-IIMV/OneDrive - IIM VIZAG/Python-Exercise-KK/geom_data/', 'India-states.csv')
ind_state_data = pd.read_csv(ind_state_filepath)
 
# Initialize the map:
m = folium.Map(location=[23, 75], zoom_start=4.25, tiles=None)
 
# Add the color for the chloropleth:
m.choropleth(
 geo_data=ind_state_basemap,
 name='choropleth',
 data=ind_state_data,
 columns=['ST_NM', 'sample_value'],
 key_on='properties.ST_NM',
 fill_color='YlGn',
 fill_opacity=0.7,
 line_opacity=0.2, legend_name='Legend')
folium.LayerControl().add_to(m)

display(m)
# Save to html
#m.save('#292_folium_chloropleth_USA1.html')
```

```python
# import libraries
import folium
import pandas as pd
 
# Make a data frame with dots to show on the map
data = pd.DataFrame({
'lat':[-58, 2, 145, 30.32, -4.03, -73.57, 36.82, -38.5],
'lon':[-34, 49, -38, 59.93, 5.33, 45.52, -1.29, -12.97],
'name':['Buenos Aires', 'Paris', 'melbourne', 'St Petersbourg', 'Abidjan', 'Montreal', 'Nairobi', 'Salvador']
})
data
 
# Make an empty map
m = folium.Map(location=[20, 0], tiles="Mapbox Bright", zoom_start=2)
 
# I can add marker one by one on the map
for i in range(0,len(data)):
    folium.Marker([data.iloc[i]['lon'], data.iloc[i]['lat']], popup=data.iloc[i]['name']).add_to(m)
 
# Save it as html
m.save('312_markers_on_folium_map1.html')
display(m)
```
