```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(200)

α_values = [0.25, 0.75]
T = 20
x = np.empty(T+1)
y = np.empty(T+1)

for α in α_values:
    x[0] = 0
    y[0] = 0
    for t in range(T):
        x[t+1] = α * x[t] + np.random.randn()
        y[t+1] = α * y[t] + α *t/2 + np.random.randn()

    plt.plot(x, label=f'No trend $\\alpha = {α}$')
    plt.plot(y, label=f'Trend $\\alpha = {α}$')

plt.legend()
plt.show()
```
