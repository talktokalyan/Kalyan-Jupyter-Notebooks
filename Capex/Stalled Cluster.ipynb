```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=  pd.read_excel(r"C:\Users\hp\Documents\Intern_IIMV\Stalled_Cost_2014-22.xlsx")
df
```

```python
df.columns =['State', 'Stalled 14-15', 'Stalled 15-16', 'Stalled 16-17','Stalled 17-18','Stalled 18-19','Stalled 19-20', 'Stalled 20-21','Stalled 21-22','Stalled 22-23']
drop = [36, 37, 38]
valid_rows = [idx for idx in drop if idx < len(df)]
df = df.drop(df.index[valid_rows])
df
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
df['Stalled 16-17'] = df['Stalled 16-17'].fillna(df['Stalled 16-17'].dropna().mean())
df['Stalled 19-20'] = df['Stalled 19-20'].fillna(df['Stalled 19-20'].dropna().mean())
df['Stalled 14-15'] = df['Stalled 14-15'].fillna(df['Stalled 14-15'].dropna().mean())
df['Stalled 18-19'] = df['Stalled 18-19'].fillna(df['Stalled 18-19'].dropna().mean())
df['Stalled 20-21'] = df['Stalled 20-21'].fillna(df['Stalled 20-21'].dropna().mean())
df['Stalled 21-22'] = df['Stalled 21-22'].fillna(df['Stalled 21-22'].dropna().mean())
df['Stalled 22-23'] = df['Stalled 22-23'].fillna(df['Stalled 22-23'].dropna().mean())
df['Stalled 17-18'] = df['Stalled 17-18'].fillna(df['Stalled 17-18'].dropna().mean())
df['Stalled 15-16'] = df['Stalled 15-16'].fillna(df['Stalled 15-16'].dropna().mean())
df.isnull().sum().sort_values(ascending = False)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
km = KMeans(n_clusters=3)
x  = df.drop("State", axis='columns')
y_predict = km.fit_predict(x)
y_predict
```

```python
df['cluster'] = y_predict
df.head()
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]
plt.figure(figsize=(8,10))
plt.scatter(df1['Stalled 14-15'],df1['State'], color ='green')
plt.scatter(df2['Stalled 14-15'],df2['State'], color ='red')
plt.scatter(df3['Stalled 14-15'],df3['State'], color ='blue')
```

```python
scaler = MinMaxScaler()
scaler.fit(df[[ 'Stalled 14-15', 'Stalled 15-16', 'Stalled 16-17','Stalled 17-18','Stalled 18-19','Stalled 19-20', 'Stalled 20-21','Stalled 21-22','Stalled 22-23']])
df[[ 'Stalled 14-15', 'Stalled 15-16', 'Stalled 16-17','Stalled 17-18','Stalled 18-19','Stalled 19-20', 'Stalled 20-21','Stalled 21-22','Stalled 22-23']] = scaler.transform(df[['Stalled 14-15', 'Stalled 15-16', 'Stalled 16-17','Stalled 17-18','Stalled 18-19','Stalled 19-20', 'Stalled 20-21','Stalled 21-22','Stalled 22-23']])
df
```
