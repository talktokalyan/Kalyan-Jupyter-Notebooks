```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 

# Folder path containing the XLSX files
folder_path = '/Users/kalyan/Library/CloudStorage/OneDrive-IIMVIZAG/Python-Exercise-KK/Kalyan-Jupyter-Notebooks/Capex/'

filename_1 = "Announced_Cost_2014-22.xlsx"
file_path_1 = os.path.join(folder_path, filename_1)

filename_2 = "Completed_Cost_2014-2022.xlsx"
file_path_2 = os.path.join(folder_path, filename_2)

df1 = pd.read_excel(file_path_1)
df2 =  pd.read_excel(file_path_2)
df1
```

```python
df1.columns =['State', 'Announced 14-15', 'Announced 15-16', 'Announced 16-17','Announced 17-18','Announced 18-19','Announced 19-20','Announced 20-21','Announced 21-22','Announced 22-23']
drop = [36, 37, 38]
valid_rows = [idx for idx in drop if idx < len(df1)]
df1 = df1.drop(df1.index[valid_rows])

df1
```

```python
df2.columns =['state', 'Completed 14-15', 'Completed 15-16', 'Completed 16-17','Completed 17-18','Completed 18-19','Completed 19-20','Completed 20-21','Completed 21-22','Completed 22-23']
drop = [36, 37, 38]
valid_rows = [idx for idx in drop if idx < len(df2)]
df2 = df2.drop(df2.index[valid_rows])

df2
```

```python
df = pd.concat([df1, df2], axis=1, join='inner')
df=df.drop('state', axis=1)
df
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
df['Completed 16-17'] = df['Completed 16-17'].fillna(df['Completed 16-17'].dropna().mean())
df['Completed 19-20'] = df['Completed 19-20'].fillna(df['Completed 19-20'].dropna().mean())
df['Completed 14-15'] = df['Completed 14-15'].fillna(df['Completed 14-15'].dropna().mean())
df['Completed 18-19'] = df['Completed 18-19'].fillna(df['Completed 18-19'].dropna().mean())
df['Completed 20-21'] = df['Completed 20-21'].fillna(df['Completed 20-21'].dropna().mean())
df['Completed 21-22'] = df['Completed 21-22'].fillna(df['Completed 21-22'].dropna().mean())
df['Completed 22-23'] = df['Completed 22-23'].fillna(df['Completed 22-23'].dropna().mean())
df['Completed 17-18'] = df['Completed 17-18'].fillna(df['Completed 17-18'].dropna().mean())
df['Completed 15-16'] = df['Completed 15-16'].fillna(df['Completed 15-16'].dropna().mean())
df.isnull().sum().sort_values(ascending = False)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
km = KMeans(n_clusters=5)
x  = df.drop("State", axis='columns')
y_predict = km.fit_predict(x)
y_predict
```

```python
df['cluster'] = y_predict
df.head()
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]
df4 = df[df.cluster == 3]
df5 = df[df.cluster == 4]
plt.figure(figsize=(8,10))
plt.scatter(df1['Announced 14-15'],df1['State'], color ='green')
plt.scatter(df2['Announced 14-15'],df2['State'], color ='red')
plt.scatter(df3['Announced 14-15'],df3['State'], color ='blue')
plt.scatter(df4['Announced 14-15'],df4['State'], color ='brown')
plt.scatter(df5['Announced 14-15'],df5['State'], color ='orange')
```
