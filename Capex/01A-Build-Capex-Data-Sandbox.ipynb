### Build Appended files from Quarterly data d/l from CMIE Capex (2014-22)

```python
# Walk through folders, sub folders and extract data from .xls and .xlsx files
# prepare appended master files 

import os
import pandas as pd
import re


folder_list = ['Completed', 'Abandoned', 'Outstanding']

for stub in folder_list:
    
    # Parent folder path
    parent_folder = f'/Users/kalyan/Library/CloudStorage/OneDrive-Personal/Kalyan/KK-Python/Kalyan-Jupyter-Notebooks/Capex/raw_data/{stub}/'
    #print('Level:0 ' + stub)
    
    # Iterate through the parent folder and subfolders
    for root, dirs, files in os.walk(parent_folder):
        #print('Level:1 ' + root)
        for subfolder in dirs:
            #print('Level:2 ' + subfolder)
            subfolder_path = os.path.join(root, subfolder)

            # Create a list to store data from .xls files in the subfolder
            data = []

            # Iterate through files in the subfolder
            for file in os.listdir(subfolder_path):
                #print('Level:3' + file)
                if file.endswith('.xls') or file.endswith('.xlsx'):
                    if file.endswith('.xls'):
                        n_skiprows = 2

                    elif file.endswith('.xlsx'):
                        n_skiprows = 1
                    file_path = os.path.join(subfolder_path, file)

                    # Extract substring from file name

                    # Regular expression pattern to extract financial year and quarter
                    pattern = r"F(\d{2}-\d{2})_Q(\d)"

                    # Extract financial year and quarter using regex
                    match = re.search(pattern, file)

                    if match:
                        financial_year = match.group(1)
                        quarter = match.group(2)
                        #print("Financial Year:", financial_year)
                        #print("Quarter:", quarter)
                    else:
                        print("No match found.")

                    # Read .xls file into a DataFrame
                    df = pd.read_excel(file_path, skiprows= n_skiprows, header=0)

                    # Add a new column with the extracted substring
                    df['FY'] = financial_year
                    df['Quarter'] = quarter
                    df['Type'] = f'{stub}'

                    # Append the DataFrame to the data list
                    data.append(df)

            # Concatenate all DataFrames in the data list
            append_df = pd.concat(data)

            # Create a separate file for each subfolder
            output_file_path = os.path.join(root, f'{stub}_{subfolder}_total.xlsx')
            append_df.to_excel(output_file_path, index=False)
            

print('FINITO')
```

#### Industry Group Concordance
Pre-requisite is Concordance file.

```python
import os
import pandas as pd

#Parent folder path
concord_folder = f'/Users/kalyan/Library/CloudStorage/OneDrive-Personal/Kalyan/KK-Python/Kalyan-Jupyter-Notebooks/Capex/raw_data/'
print('Folder:' + concord_folder)
        
df = pd.read_excel(os.path.join(concord_folder, 'Concordance-Tables-Capex.xlsx'), sheet_name = 'ind-concord')
# Count leading spaces and store in a new column
df['Leading Spaces'] = df['IndG'].str.len() - df['IndG'].str.lstrip().str.len()

df['Industry Group'] = df['IndG'].str.lstrip()

# Find the unique lowest five values in 'Leading Spaces' column
lowest_values = df['Leading Spaces'].nsmallest(100).unique()

print(lowest_values)
# Create separate columns for each of the values
for value in lowest_values:
    column_name = f'IndG_L{value}'
    df[column_name] = df.apply(lambda row: row['Industry Group'] if row['Leading Spaces'] == value else None, axis=1)
    df[column_name] = df[column_name].ffill().str.lstrip()


# Folder path containing the XLSX files
out_folder_path = '/Users/kalyan/Library/CloudStorage/OneDrive-Personal/Kalyan/KK-Python/Kalyan-Jupyter-Notebooks/Capex/proc_data/'

# Output file path and name
out_file_name = 'Industry-group-hierarchy.xlsx'
out_file_path = os.path.join(out_folder_path, out_file_name)

df.to_excel((out_file_path), index=False)

print('Finito')
```
