```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import pandas as pd
from plotly import tools
import plotly.express as px
import os 

# Folder path containing the XLSX files
in_folder_path = '/Users/kalyan/Library/CloudStorage/OneDrive-IIMVIZAG/Python-Exercise-KK/Kalyan-Jupyter-Notebooks/Capex/raw_data/'

# filename_1 = "Project_CapEx_Final.xlsx"
filename_1 = "RoughBook.xlsx"

file_path_1 = os.path.join(in_folder_path, filename_1)

df = pd.read_excel(file_path_1 , sheet_name='Events')

# Filter rows with 'Announced' and 'Completed' status
announced_rows = df[df['Status'] == 'Announced']
completed_rows = df[df['Status'] == 'Completed']

# Merge announced and completed rows on 'Name' column
merged = pd.merge(announced_rows, completed_rows, on='Name', suffixes=('_announced', '_completed'))

# Calculate elapsed time
merged['Elapsed_Time'] = merged['Date_completed'] - merged['Date_announced']

print(merged[['Name', 'Elapsed_Time']])

print(df)
print(df.columns)
print(df.to_json())
```
