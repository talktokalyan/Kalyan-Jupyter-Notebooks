```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(200)

α_values = [0.25, 0.5]
T = 200
x = np.empty(T+1)
y = np.empty(T+1)

for α in α_values:
    x[0] = 0
    y[0] = 0
    for t in range(T):
        x[t+1] = α * x[t] + np.random.randn()
        y[t+1] = α * y[t] + α *t/2 + np.random.randn()

    plt.plot(x, label=f'No trend $\\alpha = {α}$')
    plt.plot(y, label=f'Trend $\\alpha = {α}$')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(456)

α_values = [0.05]
T = 200
a = np.empty(T+1)
b = np.empty(T+1)
c = np.empty(T+1)
d = np.empty(T+1)
e = np.empty(T+1)

for α in α_values:
    a[0] = 1
    b[0] = 1
    c[0] = 1
    d[0] = 1
    e[0] = 1
    for t in range(T):
        a[t+1] = a[t] + np.random.randn() # Random Walk No Drift
        b[t+1] = 0.025 + b[t] + np.random.randn() # Random Walk with Drift
        c[t+1] = c[t] + α*t/200 + np.random.randn() # Random Walk with trend
        d[t+1] = α*t/200 + np.random.randn() # Trend Stationary Process 
        e[t+1] = 0.9*e[t] + α*t/200 + np.random.randn() # Non Unit root AR1 Process + trend

plt.plot(a, label=f'a - RNW No Drift ')
plt.plot(b, label=f'b - RNW with Drift')
plt.plot(c, label=f'c - RNW + Trend $\\alpha = {α}t$')
plt.plot(d, label=f'd - Trend $\\alpha = {α}t$')
plt.plot(e, label=f'e - Non Unit root AR(1) + Trend $\\alpha = {α}t$')

plt.legend()
plt.show()
```
